```python
import pandas as pd
import numpy as np
```

```python

n_members = 300
members = [f"M{i:03d}" for i in range(1, n_members + 1)]
groups = np.random.choice(['A', 'B'], size=n_members)

records = []
start_date = pd.to_datetime("2025-03-01")
for member, group in zip(members, groups):
    for week in range(1, 13):
        due_date = start_date + pd.Timedelta(weeks=week)
        delay = int(np.random.normal(loc=0 if group == 'A' else 2, scale=2.5))
        payment_date = due_date + pd.Timedelta(days=delay)
        amount = round(np.random.normal(loc=3500 if group == 'A' else 3300, scale=300), 2)
        amount = max(amount, 0)
        days_late = (payment_date - due_date).days
        is_late = days_late > 0
        records.append([member, group, due_date, payment_date, days_late, is_late, amount])

df_ab = pd.DataFrame(records, columns=[
    "member_id", "group", "due_date", "payment_date", 
    "days_late", "is_late", "amount"
])
# Save the DataFrame to a CSV file
df_ab.to_csv("members_ab.csv", index=False)
```

```python
group_stats = df_ab.groupby("group").agg({
    "days_late": ["mean", "median"],
    "is_late": ["mean", "sum"],
    "amount": ["mean", "sum", "count"]
}).round(2)
```

```python
group_stats.columns = ['_'.join(col).strip() for col in group_stats.columns.values]
group_stats = group_stats.reset_index()
group_stats
```
